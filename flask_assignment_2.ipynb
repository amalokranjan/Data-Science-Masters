### Q1. Explain GET and POST methods.
    GET and POST are HTTP methods used in web development to request or submit data to a server.
    
    GET: It is a method that retrieves or requests data from a specified resource. It is used to retrieve data that has been cached or saved previously, and it is also useful for displaying data, such as web page, on a browser. The data sent in a GET request is appended to the URL in the form of a query string. It is generally not recommended to use GET requests for submitting sensetive information as the data is visible in the URL.
    
    POST: It is a method that submits data to be processed to a specific resource. POST requests are used to send data to the server to be processed, such as a login form or a comment submission form. The data sent in a POST request is sent in the body of the request and is not visible in the URL. It is the preferred method for submitting sensitive information.

### Q2. Why is request used in Flask?
    The request object in Flask is used to access incoming request data in Flask applications. It provides a way to access form data, cookies, query parameters, and files uploaded by the client. By using the request object, Flask applications can extract information from incoming requests and use it to generate responses. For example, when a user submits a form on a web page, the form data can be extracted from the request object and processed by the Flask application to produce a response. The request object is an instance of the Request class in the Werkzeug library, which is the underlying HTTP library used by Flask.

### Q3. Why is redirect() used in Flask?
    redirect() function in Flask is used to redirect the user to a different endpoint. It is used when a user is accessing a particular endpoint, but it is necessary to redirect them to another endpoint. This function is commonly used after a form submission or after a user logs in successfully.
    
    When a user submits a form on a webpage, the data is sent to a Flask endpoint via a POST request. After the form data is processed, it might be necessary to redirect the user to a different page (e.g - a thank you page or a page showing the results of their form submission). This is where the redirect() function comes into play.

### Q4. What are templates in Flask? Why is the render_template() function used?
    In Flask, templates are used to dynamically generate HTML pages by using placeholders for dynamic content. These templates are typically written in HTML and may contain placeholders, conditional statements, and loops that are replaced with actual values during runtime.
    
    The render_template() function is used to render a template and return the generated HTML. It takes at least one argument, which is the name of the template file, and any additional arguments that are needed to render the template. The function searches for the template file in the templates directory by default.

```python
from flask import Flask, request

app = Flask(__name__)

@app.route('/square', methods=['POST'])

def square():
    data = request.get_json()
    number = data['number']
    result = number ** 2
    return {'result': result}

if __name__ == '__main__':
    app.run("0.0.0.0")
```
