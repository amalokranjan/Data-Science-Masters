### Q1. What is Flask Framework? What are the advantages of Flask Framework?
    Flask is a web framework written in Python used for building web applications. It si a micro-framework and provides the basic functionality for building web applications. Flask is designed to be very flexible and modular, making it easy to customize and extend to fit the specific needs of an application.
    
    Some advantages of Flask framework are:
    
    1. Lightweight: Flask is lightweight framework that requires minimal setup and configuration, making it easy to use for small and medium-sized projects.
    2. Easy to use: Flask is easy to learn and use, with a simple and intuitive syntax that allows developers to quickly build web applications.
    3. Modular: Flask is highly modular, with a wide range of extensions an plugins available that can be used to add additional functionality to the framework.
    4. Flexible: Flask is designed to be flexible and can be used for a wide range of applications, from small personal projects to large enterprise applications.
    5. Powerful: Flask provides powerful tools and features, such as built-in support for Jinja2 templates and the ability to easily integrate with other Python libraries and tools.

### Q3. What is App routing in Flask? Why do we use app routes?
    In Flask, routing refers to determining how the application responds to a client request to a specific endpoint, which is a specific URL requested by the client. App routing in Flask refers to the process of defining the URLs that the application can handle and mapping those URLs to appropriate view functions.
    App routes are used to define the request URLs that the Flask application can handle. They are defined using the @app.route() decorator and the funciton that handles the request is associated with defined route. The decorator accepts the URL pattern as its argument and associates the function with the URL pattern.

### Q4. Create a "/welcome" route to display the welcome message "Welcome to ABC Corporation" and a "/" route to show the following details:
    Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999
Attach the screenshot of the output in Jupyter Notebook

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def fun1():
    company_name = "ABC Corporation"
    location = "India"
    contact_detail = "999-999-9999"
    
    return "Company Name: {}\nLocation: {}\nContact Detail: {}".format(company_name, location, contact_detail)


@app.route('/welcome')
def fun2():
    return "Welcome to ABC Corporation"

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```

### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
    The url_for() function is used in Flask for URL building. It generates a URL to the specified endpoint with the method provided.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello World'

@app.route('/user/<username>')
def profile(username):
    return f'{username}\'s profile'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('profile', username='alok'))
    
if __name__ == '__main__':
    app.run(host="0.0.0.0")
```
